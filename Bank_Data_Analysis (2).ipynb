# 2. Data Cleaning

Check for missing values (nulls).

```python
df.isnull().sum()
```

```python
unknown_counts = (df == "Unknown").sum()
unknown_counts[unknown_counts > 0]
```

This shows while the data doesn't contain any nulls, it contains "Unknown" which is similar as null but won't appear in the null check

```python
for col in categorical_cols:
    if "Unknown" in df[col].unique():
        df[col] = df[col].replace("Unknown", df[col].mode()[0])
```

"Unknown" is only shown in categorical columns so we only need to replace the unknowns with the mode

```python
(df == "Unknown").sum().sum()
```

```python
df.duplicated().sum()
```

This verifies we have successfully removed any unknowns from the data

```python
print("Age range:", df["Customer_Age"].min(), "-", df["Customer_Age"].max())
print("Utilization ratio range:", df["Avg_Utilization_Ratio"].min(), "-", df["Avg_Utilization_Ratio"].max())
```

Age and ultilization ratio are all within its expected resonable range

```python
#histogram for customer age
plt.figure(figsize=(8,5))
sns.histplot(df["Customer_Age"], bins=20)
plt.xlabel("Customer Age")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Age")
plt.show()
```

This shows us that the most bank churners tend to be around the ages of 40-50

This shows that the majority of the customers have a blue card and then its silver then gold and finally platinum

## Bivariate Analysis


```python
# calculate % churn by gender
churn_by_gender = (
    df.groupby("Gender")["Attrition_Flag"]
      .value_counts(normalize=True)   # gives percentages
      .rename("percentage")
      .reset_index()
)

# plot
plt.figure(figsize=(6,4))
sns.barplot(data=churn_by_gender, x="Gender", y="percentage", hue="Attrition_Flag")
plt.title("Percentage of Churn by Gender")
plt.ylabel("Percentage")
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))  # show % format
plt.show()
```

This shows that the bank has almost equal percentages of existing customers being male and female with male being slightly higher however however the attired customers also are almost equal but the female attrited customer seems to be slightly higher

```python
# calculate percentage churn by education level
edu_churn = (
    df.groupby("Education_Level")["Attrition_Flag"]
      .value_counts(normalize=True)
      .rename("percentage")
      .reset_index()
)

# plot percentages
plt.figure(figsize=(8,5))
sns.barplot(data=edu_churn, x="Education_Level", y="percentage", hue="Attrition_Flag",
            order=df["Education_Level"].value_counts().index)
plt.title("Percentage of Churn by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Percentage")
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))  # show as %
plt.show()
```

The percentage of existing customers between the education level of uneducated to graduate level are on almost the same percentage of existing and attrited customer

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Attrition_Flag", y="Total_Trans_Ct")
plt.title("Total Transaction Count vs Attrition")
plt.show()
```

This confirms that customers with lower transaction counts are more likely to leave the bank

```python
# create age groups
bins = [0, 29, 39, 49, 59, 100]
labels = ["<30", "30-39", "40-49", "50-59", "60+"]
df["Age_Group"] = pd.cut(df["Customer_Age"], bins=bins, labels=labels)

# plot churn %
age_churn = (
    df.groupby("Age_Group")["Attrition_Flag"]
      .value_counts(normalize=True)
      .rename("percentage")
      .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=age_churn, x="Age_Group", y="percentage", hue="Attrition_Flag")
plt.title("Churn by Age Group")
plt.ylabel("Percentage")
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))
plt.show()
```

This shows that the age group under 30 has the highest percentage of existing customers and that it follows a general trend of as the age increases the percentage of attired customers increases except at the age of 60+ where it slightly falls to around the same percentage as the age group 30-39

This shows that marriage status has no effect on if the customer stays or leaves the bank, as the percentage of existing customers are all almost identical and same with percentage of attrited customers

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Attrition_Flag", y="Months_Inactive_12_mon")
plt.title("Months Inactive (last 12 months) vs Attrition")
plt.xlabel("Attrition Flag")
plt.ylabel("Months Inactive (12 months)")
plt.show()
```

This shows that as the Months Inactive increases, the number of Attrited Customers also increases (Positive correlation).

```python
# calculate percentage churn by dependent count
dep_churn = (
    df.groupby("Dependent_count")["Attrition_Flag"]
      .value_counts(normalize=True)
      .rename("percentage")
      .reset_index()
)

# plot percentages
plt.figure(figsize=(8,5))
sns.barplot(data=dep_churn, x="Dependent_count", y="percentage", hue="Attrition_Flag")
plt.title("Churn by Dependent Count")
plt.xlabel("Number of Dependents")
plt.ylabel("Percentage")
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))  # show %
plt.show()
```

This shows that the number of depends has almost no effect if the customer stays or leaves

```python
# pivot for stacked plot
income_churn_pivot = income_churn.pivot(index="Income_Category", columns="Attrition_Flag", values="percentage")

income_churn_pivot.plot(kind="barh", stacked=True, figsize=(8,5), colormap="Set2")
plt.title("Stacked Churn Percentage by Income Category")
plt.xlabel("Percentage")
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))
plt.show()
```

All income categories seem to have a similar attrited and existing customers, the under 40k and 120k+ seem to have slightly more attrited customers compared to the other categories

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Attrition_Flag", y="Avg_Utilization_Ratio")
plt.title("Avg Utilization Ratio vs Attrition")
plt.show()
```

This shows that the attrited customer has much lower avg utilization ratio than the existing customer
